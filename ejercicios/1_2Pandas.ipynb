# pd.pivot_table


```python
def moda(x):
  return x.mode().iloc[0] if not x.mode().empty else None

# Precio promedio unitario de cada país y descripción
pivot_table_1= pd.pivot_table(data=df_ventas,
                              values="UnitPrice",
                              index="Country",
                              columns="Description",
                              aggfunc= moda)


pivot_description_Country= df_ventas.pivot_table(values="UnitPrice",
                              index="Country",
                              columns="Description",
                              aggfunc= moda)
pivot_description_Country
```

# pd.to_datetime

```python
df_ventas["month"] =pd.to_datetime(df_ventas["InvoiceDate"]).dt.month
df_ventas.head()
```

```python
#Total de ventas por país y por mes

pivot_country_month_quantity= df_ventas.pivot_table(values= "Quantity",
                                                     index=("Country", "Description"),
                                                     columns="month",
                                                     aggfunc="sum") 

pivot_country_month_quantity
```

# melt

```python
df_melt_quantity_price= df_ventas.melt(
  id_vars= ['InvoiceNo', 'StockCode', 'Description', 'InvoiceDate','CustomerID', 'Country', 'month'],
  value_vars= ['Quantity','UnitPrice','Discount' ],
  var_name= "Métrica" ,
  value_name= "Valor")

df_melt_quantity_price
```
