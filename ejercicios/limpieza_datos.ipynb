````python
df.head()
````

### Cambio tipo de datos

````python
meses = {
    'enero': '01',
    'febrero': '02',
    'marzo': '03',
    'abril': '04',
    'mayo': '05',
    'junio': '06',
    'julio': '07',
    'agosto': '08',
    'septiembre': '09',
    'octubre': '10',
    'noviembre': '11',
    'diciembre': '12'
}
````

````python
df['Invoice_Date'] = df['Invoice_Date'].replace(meses, regex=True)

````

````python
df.head()
````

````python
df['Invoice_Date']=pd.to_datetime(df['Invoice_Date'], format= "%Y-%m-%d %H:%M")
````

````python
str_float=['Unit_Price', 'Shipping_Cost']
for col in str_float:
   df[col]=df[col].str.replace(',', '.')
  
````

````python
for col in str_float:
 df[col]=df[col].apply(lambda x:float(x))
````

````python
diccionario_map={1:'returned', 0:'not returned'}
df['Return_Status']=df['Return_Status'].map(diccionario_map)
````

````python
df['Year']= df['Invoice_Date'].dt.year
df['Month']=df['Invoice_Date'].dt.month
````

````python
df['Day_name']=df['Invoice_Date'].dt.day_name()
df.head()
````

````python
df['Week']=df['Invoice_Date'].dt.isocalendar().week
````

````python
df['Quarter']=df['Invoice_Date'].dt.quarter
````

````python
df.to_csv('data/online_sales_limpio.csv')
````
