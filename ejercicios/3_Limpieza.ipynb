### 1.2. Propósito del reto e información de los datos.

El propósito de este reto guiado es poner en práctica los conocimientos de limpieza de datos para poder realizar un buen posterior análisis exploratorio de los datos.

El objetivo de hoy es limpiar los datos de los archivos. Estos coonjuntos de datos están relacionados con campañas de *marketing* directo de una institución bancaria portuguesa. Las campañas de marketing se basaron en llamadas telefónicas. A menudo, se requería más de un contacto con el mismo cliente para determinar si el producto (depósito a plazo bancario) sería suscrito o no. Las colummas que tenemos en el primer *dataset* ('bank-additional.csv') son:  


- **age**: La edad del cliente.

- **job**: La ocupación o profesión del cliente.

- **marital**: El estado civil del cliente.

- **education**: El nivel educativo del cliente.

- **default**: Indica si el cliente tiene algún historial de incumplimiento de pagos (1: Sí, 0: No).

- **housing**: Indica si el cliente tiene un préstamo hipotecario (1: Sí, 0: No).

- **loan**: Indica si el cliente tiene algún otro tipo de préstamo (1: Sí, 0: No).

- **contact**: El método de contacto utilizado para comunicarse con el cliente.

- **duration**: La duración en segundos de la última interacción con el cliente.

- **campaign**: El número de contactos realizados durante esta campaña para este cliente.

- **pdays**: Número de días que han pasado desde la última vez que se contactó con el cliente durante esta campaña.

- **previous**: Número de veces que se ha contactado con el cliente antes de esta campaña.

- **poutcome**: Resultado de la campaña de marketing anterior.

- **emp.var.rate**: La tasa de variación del empleo.

- **cons.price.idx**: El índice de precios al consumidor.

- **cons.conf.idx**: El índice de confianza del consumidor.

- **euribor3m**: La tasa de interés de referencia a tres meses.

- **nr.employed**: El número de empleados.

- **y**: Indica si el cliente ha suscrito un producto o servicio (Sí/No).

- **date**: La fecha en la que se realizó la interacción con el cliente.

- **contact_month**: Mes en el que se realizó la interacción con el cliente durante la campaña de marketing.

- **contact_year**: Año en el que se realizó la interacción con el cliente durante la campaña de marketing.

- **id_**: Un identificador único para cada registro en el dataset.

El segundo set de datos ('customer-details.xlsx') es un excel que nos da información sobre las características demográficas y comportamiento de compra de los clientes del banco. Este excel consta de 3 hojas de trabajo diferentes, en cada una de ellas tenemos los clientes que entraron en el banco en diferentes años. Sus columnas son:

- **Income**: Representa el ingreso anual del cliente en términos monetarios.

- **Kidhome**: Indica el número de niños en el hogar del cliente.

- **Teenhome**: Indica el número de adolescentes en el hogar del cliente.

- **Dt_Customer**: Representa la fecha en que el cliente se convirtió en cliente de la empresa.

- **NumWebVisitsMonth**: Indica la cantidad de visitas mensuales del cliente al sitio web de la empresa.

- **ID**: Identificador único del cliente.

```python
### 2.3. Unión de datos
###Une los datos de todos los DataFrames. Primero los del archivo customer-details con un concat() y posteriormente con el DataFrame del ejercicio 1 con un merge()

customer = pd.concat([customer_2012,customer_2013,customer_2014], axis=0, ignore_index=True)
```

```python
customer.describe().T
```

```python
customer.duplicated().sum()
```

```python
bank_final= bank.merge(customer, how= 'inner',left_on = 'id_', right_on = "ID")
```

```python
### 2.4. Limpieza de los datos.
#2.4.1 Elimina columnas redundantes. 
#- Para ello primero identifica estas columnas y utiliza el método drop. Recuerda que debes especificar el eje en el que vamos a hacer la eliminación y que si queremos sobreescribir el DataFrame habría que usar inplace=True
bank_final.drop('id_', axis=1, inplace=True)
```

```python
#2.4.2 Homogeneiza los nombres de las columnas.
#- Primero recordemos las columnas que tenemos usando columns
# Si nos fijamos bien vemos cómo hay columnas con mayusculas, con minúsculas, sin espacios, etc. Plantéate una forma de homogeneizar estas columnas
# Crea primero un diccionario con los nombres antiguos de las columnas y con los nuevos
# Utiliza rename para renombrar el DataFrame.

nombresColumnas= bank_final.columns
```

```python
renombre_columnas = {columna: columna.lower().replace('.', '_') for columna in nombresColumnas}
renombre_columnas
```

```python
bank_final.rename(columns= renombre_columnas, inplace=True)
bank_final.columns
```

```python
#2.4.3 Haz que todos los datos estén en minúsculas. Para ello utiliza .str.lower()
for col in bank.columns.to_list():
  try:
    bank[col] = bank[col].str.lower()
  except:
    pass
  
```

```python
#2.4.4 Crea nuevas columnas para el mes('contact_month') y el año('contact_year') de la columna 'date'. 
# Utiliza str.split() para indicar en qué parte del srting quieres cortar
# Luego selecciona las posiciones de la lista que nos interesa 

fecha = bank_final['date'].str.split("-", expand=True)
fecha
```

```python
bank_final['contact_year'] = fecha[2]
bank_final['contact_month'] = fecha[1]
```

```python
# 2.4.5 Convierte la columna 'date' a tipo datetime.
# Primero crea un diccionario con los nombres de los meses y el número del mes.
meses = {
    "enero": '01',
    "febrero": '02',
    "marzo": '03',
    "abril": '04',
    "mayo": '05',
    "junio": '06',
    "julio": '07',
    "agosto": '08',
    "septiembre": '09',
    "octubre": '10',
    "noviembre": '11',
    "diciembre": '12'
}

# Ahora reemplaza los datos de 'date' usando .replace()


# Por último usa pd.todatetime para cambiar el tipo de dato de la columna (Recuerda usar el formato %d-%m-%Y)
```

```python
bank_final.replace({'date':meses}, regex = True,inplace = True)
bank_final.head()
```

```python
bank_final["date"]= pd.to_datetime(bank_final["date"], format= '%d-%m-%Y')
bank_final.info()
```

```python
# 2.4.6 Reemplaza los . por espacios en la columna 'education'. Utiliza.str.replace()
bank_final['education']= bank_final['education'].str.replace("."," ")
bank_final.head()
```

```python
bank_final.columns
```

```python
#2.4.7 Convierte a float las columnas que deberían serlo
# Primero identifica cuáles son estas columnas y crea una lista con los nombres de estas columnas.
# Ahora itera en esa lista con un bucle for.
# Dentro del bucle utiliza str.replace() para cambiar las ',' por '.'.
# Aún dentro del bucle utiliza un apply con una función lambda en la que cambiemos el tipo de dato a float.
columnas_str_float= ['cons_price_idx', 'cons_conf_idx', 'euribor3m','nr_employed']

for columna in columnas_str_float:
  bank_final[columna]= bank_final[columna].str.replace(",",".").astype(float)
  
```

```python
#2.4.8 Mapear los valores 0 y 1 por No y Sí.
# Creamos un diccionario con claves 0 y 1 y valores 'No' y 'Sí'.
# Crea una lista con las columnas que vamos a mapear
# Itera con un bucle for por la lista y usa el método map() para cada columna
diccionario_booleanos ={0:'no', 1:'si'}
columnas_bool= ['default', 'housing', 'loan']
for columna in columnas_bool:
  bank_final[columna]=  bank_final[columna].map(diccionario_booleanos)
```
