```python
%pip install pdfplumber
import pdfplumber as pdfplu
import pandas as pd
```

```python


# Abre el archivo PDF
with pdfplu.open(r"C:\Users\maria\Downloads\0_7529_1.pdf") as pdf:
    # Selecciona la primera pgina (puedes iterar para leer ms pginas)
    pagina = pdf.pages[0]
    
    # Extrae las tablas de la pgina
    tablas = pagina.extract_tables()

    # Convierte la tabla en un DataFrame de pandas
    if tablas:
        for i, tabla in enumerate(tablas):
            df = pd.DataFrame(tabla)
            print(f"Tabla {i}:\n", df)
    else:
        print("No se encontraron tablas en la pgina.")
```
