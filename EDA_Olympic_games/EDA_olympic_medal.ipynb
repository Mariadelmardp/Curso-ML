Evolucion de las disciplinas  a lo largo de la historia. 

```python
df_medals_discipline=medals.groupby('slug_game')['discipline_title'].count().reset_index()
df_medals_discipline['year']=df_medals_discipline['slug_game'].str.split('-').str[-1]
df_medals_discipline=df_medals_discipline.sort_values(by='year')
df_medals_discipline
```

## El porcentaje de disciplinas a lo largo de la Historia de los Juegos Olimpicos a crecido un 842.86

```python
Crecimiento=round((df_medals_discipline['discipline_title'].iloc[-2]-df_medals_discipline['discipline_title'].iloc[0])/df_medals_discipline['discipline_title'].iloc[0]*100,2)
```

```python
print(f"El porcentaje de disciplinas a lo largo de la Historia de los Juegos Olimpicos a crecido un {Crecimiento }")
```
