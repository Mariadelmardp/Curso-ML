- **`discipline_title`**: Título de la disciplina deportiva (por ejemplo, atletismo, natación).

- **`event_title`**: Nombre del evento específico dentro de la disciplina (por ejemplo, 100m libres, salto de altura).

- **`slug_game`**: Identificador único del juego o competición, usado en URL y para fines de indexación.

- **`participant_type`**: Tipo de participante en el evento (puede ser individual o en equipo).

- **`medal_type`**: Tipo de medalla obtenida (oro, plata, bronce).

- **`athletes`**: Lista o número de atletas que participan en el evento.

- **`rank_equal`**: Indicador de posición compartida (si dos o más participantes empatan en la misma posición).

- **`rank_position`**: Posición o rango alcanzado por el participante en el evento.

- **`country_name`**: Nombre completo del país del participante.

- **`country_code`**: Código de dos letras que representa al país (por ejemplo, US para Estados Unidos).

- **`country_3_letter_code`**: Código de tres letras del país (por ejemplo, USA para Estados Unidos).

- **`athlete_url`**: URL del perfil o página del atleta.

- **`athlete_full_name`**: Nombre completo del atleta.

- **`value_unit`**: Unidad de medida del valor registrado en la competencia (por ejemplo, segundos, metros).

- **`value_type`**: Tipo de valor registrado (por ejemplo, tiempo, distancia).
